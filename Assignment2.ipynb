### 문제 1. 로그인 시스템 제어 (while & if-else)

사용자의 `id`와 `pw`를 입력받아 로그인을 처리하는 로직을 작성하세요.

- 시스템에 등록된 정보: `stored_id = "admin"`, `stored_pw = "pass123"`
- 사용자가 정확한 정보를 입력할 때까지 **무한히 반복**해서 입력을 받습니다.
- 로그인에 성공하면 `"환영합니다, 관리자님!"`을 출력하고 반복문을 탈출합니다.
- 로그인에 실패하면 `"아이디 또는 비밀번호가 틀렸습니다. 다시 시도하세요."`를 출력합니다.

```python
stored_id = "admin"
stored_pw = "pass123"
# 여기에 코드를 작성하세요.
```

````python
stored_id = "admin"
stored_pw = "pass123"

while True:
    id_input = input("id 를 입력하세요")
    pw_input = input("pw 를 입력하세요")
    
    if id_input==stored_id and pw_input==stored_pw:
        print("환영합니다, 관리자님!")
        break
        
    print("아이디 또는 비밀번호가 틀렸습니다. 다시 시도하세요.")
````

### 문제 2. 3-6-9 게임기 (for & range & if)

1부터 30까지의 숫자를 출력하되, 숫자에 **3, 6, 9**가 포함되어 있는 경우 숫자 대신 "짝!"을 출력하는 코드를 작성하세요.

- 팁: 숫자를 문자열로 변환하여 '3', '6', '9'가 들어있는지 확인(in 연산자)하면 쉽습니다.

```python
# 여기에 코드를 작성하세요.
```

````python
# 짝을 여러번

# Anti pattern - 나중에 수정
stringify_num = lambda x: list(str(x))

list_b = ["3", "6" , "9"]

for num in range(1, 31):
    
    list_a = stringify_num(num)
    
    shared = [target for target in list_a if target in list_b]
    if not shared:
        print(num)
    else:
        print("짝" * len(shared)) 
````

### 문제 3. 장바구니 품목 필터링 및 할인 적용 (Comprehension)

다음은 장바구니에 담긴 상품 리스트입니다. 

가격이 **50,000원 이상인 상품만 골라 10% 할인**을 적용한 새로운 딕셔너리 `discounted_cart`를 만드세요.

- 결과 예시: `{'Laptop': 900000.0, 'Monitor': 180000.0}`
- 반드시 **딕셔너리 컴프리헨션**을 사용하여 한 줄로 작성하세요.

```python
cart = {"Laptop": 1000000, "Mouse": 15000, "Monitor": 200000, "Pad": 5000}
# 여기에 코드를 작성하세요.
```

````python
cart = {"Laptop": 1000000, "Mouse": 15000, "Monitor": 200000, "Pad": 5000}
criteria_price = 50000

discounted_cart = {item: cart[item] * 0.9 for item in cart if cart[item] >= criteria_price}

print(discounted_cart)
````

### 문제 4. 구구단 가로 출력 (Nested Loop)

중첩 `for`문을 사용하여 구구단 2단부터 9단까지 출력하세요. 단, 각 단은 아래 예시처럼 **가로로 한 줄씩** 출력되어야 합니다.

- 출력 예시:
    
    `2*1=2 2*2=4 ... 2*9=18`
    
    `3*1=3 3*2=6 ... 3*9=27`
    

```python
# 여기에 코드를 작성하세요.
```

````python
def head(k ,v):
    return f"{k}*{v}={k*v}"

def oneLiner(line_no): 
    line_array = []
    for i in range(1,10):
        line_array.append(head(line_no , i))
    seperator = " "
    seperator.join(line_array)
    print(seperator.join(line_array))
 
 
for line in range(2,10):
    oneLiner(line) 
     
     
````


### 문제 5. 데이터 구조 재구성 (zip & Unpacking)

웹사이트 방문자 데이터가 `[(날짜, 방문자수), ...]` 형태의 튜플 리스트로 저장되어 있습니다. 이 데이터를 시각화하기 위해 날짜 리스트(`dates`)와 방문자수 리스트(`counts`)로 각각 분리(Unzipping)한 뒤, 방문자 수의 **평균**을 계산하세요.

```python
visitor_data = [("2026-03-01", 120), ("2026-03-02", 150), ("2026-03-03", 90), ("2026-03-04", 200)]
# 여기에 코드를 작성하세요.
```

````python
visitor_data = [("2026-03-01", 120), ("2026-03-02", 150), ("2026-03-03", 90), ("2026-03-04", 200)]

# Method 1 (dirty) 
date_array = []

for dates in visitor_data:
    date_array.append(dates[1])
 
avg = sum(date_array)/len(date_array)
print(avg)

# Method 2 
dates, visitors = zip(*visitor_data)
print(sum(visitors)/len(visitors))
````
